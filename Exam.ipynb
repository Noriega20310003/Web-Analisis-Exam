```python
from urllib import request
import re
import pandas as pd
from bs4 import BeautifulSoup
from PyPDF2 import PdfReader, PdfWriter
```

```python
hdr = {
    'User-Agent': 'Wget/1.14 (linux-gnu)',
    'Accept':'*/*'
}
```

```python
# CONAHCYT
conahcyt = "https://centrosconahcyt.net/oferta-academica/"
req = request.Request(conahcyt, headers=hdr)
html_conahcyt = request.urlopen(req).read().decode('utf8')
data_conahcyt = BeautifulSoup(html_conahcyt, 'html.parser')

carreras_divs = data_conahcyt.find_all("div",{"class": "blog-masonry-item"})

ofertas_academicas = {}

for div in carreras_divs:
    sede = div.find('span').text
    carrera = div.find('h4').text
    area = div.find(class_='post-meta').text
    area = re.sub("\n", "", area)

    ofertas_academicas[carrera] = [sede, area]

del ofertas_academicas['Cursos - CIATEC']
del ofertas_academicas['CIATEQ']
del ofertas_academicas['GCI']
del ofertas_academicas['JUVE']

df = pd.DataFrame.from_dict(ofertas_academicas, orient='index', columns=['Sede', 'Área de Investigación'])
df.reset_index(inplace=True)
df.rename(columns={'index': 'Carrera'}, inplace=True)

df.to_excel('ofertas_academicas_conahcyt.xlsx', index=False)

df
```

```python
# escuela TEC de monterrey
escuela_priv_1 = "https://maestriasydiplomados.tec.mx/programas/posgrados"
req = request.Request(escuela_priv_1, headers=hdr)
html_escuela_priv_1 = request.urlopen(req).read().decode('utf8')
data_priv_1 = BeautifulSoup(html_escuela_priv_1, 'html.parser')

carreras_priv_1 = data_priv_1.find_all("div",{"class": "newplcol"})

carreras_escuela_privada_1 = []

for div in carreras_priv_1:
    carrera = div.find('h3').text
    carreras_escuela_privada_1.append(carrera)

df_priv_1 = pd.DataFrame(carreras_escuela_privada_1)
df_priv_1.columns = ["Carrera"]
df_priv_1.to_excel("ofertas_escuela_privada_1.xlsx", index=False)
df_priv_1
```

```python
# Poner el articulo o libro aquí en pdf
pdf_path = 'cadena-de-custodia_guia-nacional.pdf'
pdf = PdfReader(pdf_path)
```

```python
pattern = input("Ingresa la palabra o frase a buscar: ")

pdf_resultado = PdfWriter()

for num_pagina, pagina in enumerate(pdf.pages, start=1):
    texto = pagina.extract_text()

    coincidencias = re.findall(pattern, texto)

    if coincidencias:
        pdf_resultado.add_page(pagina)

        print(f'Página {num_pagina} con coincidencias:')
        print(texto)

pdf_resultado_path = 'resultado_busqueda_punto3.pdf'
with open(pdf_resultado_path, 'wb') as pdf_final:
    pdf_resultado.write(pdf_final)
```

```python
# EBAY
ebay = "https://www.ebay.com/globaldeals?_trkparms=pageci%3A14b7de28-6fd7-11ee-8b21-726181175aa4%7Cparentrq%3A50d1be8318b0a74460b2d565fffd9e07%7Ciid%3A1"
req = request.Request(ebay, headers=hdr)
html_ebay = request.urlopen(req).read().decode('utf8')
data_ebay = BeautifulSoup(html_ebay, 'html.parser')
```

```python
productos_divs = data_ebay.find_all("div",{"class": "col"})

productos = {}

for div in productos_divs:
    producto_div = div.find("span",{"itemprop": "name"})
    producto = re.findall(r'\"\>.*\<\/', str(producto_div))
    str_producto = "".join(producto)
    str_producto = re.sub("\"\>", "", str(str_producto))
    str_producto = re.sub("\<\/", "", str(str_producto))


    precio_div = div.find("span",{"itemprop": "price"})
    precio = re.findall(r'\$\d+\.\d{2}', str(precio_div))
    str_precio = "".join(precio)

    productos[str_producto] = [str_precio]

df = pd.DataFrame.from_dict(productos, orient='index', columns=['Precio'])
df.reset_index(inplace=True)
df.rename(columns={'index': 'Producto'}, inplace=True)

df.to_excel('productos_tienda1.xlsx', index=False)

df
```

```python
# odessalibrerias
odessa = "https://www.odessalibrerias.com.mx/"
data = BeautifulSoup(request.urlopen(request.Request(odessa, headers=hdr)).read().decode('utf8'), 'html.parser')
productos = data.find_all("div",{"class": "product"})
productos1 = {}
for div in productos:
    productos1[div.find('h3').text] = ["".join(div.find("span",{"class": "price"}))]
df = pd.DataFrame.from_dict(productos1, orient='index', columns=['Precio'])
df.reset_index(inplace=True)
df.rename(columns={'index': 'Producto'}, inplace=True)
df.to_excel('productosOdessa.xlsx', index=False)
df
```
